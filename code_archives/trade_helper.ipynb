```python
import time
import random
import pandas as pd
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import warnings
warnings.filterwarnings('ignore')
```

## Loading all my cards

### functions

```python
class Scrapper:
    
    _login = "jordy.bonnet@gmail.com"
    _mdp = "Mim0ur4$ofia"

    def __init__(self, url):
        self.url = url
        self.service = Service(ChromeDriverManager().install())
        self.driver = webdriver.Chrome(service=self.service)
        self.QUICK_WAIT = [1, 3]
        self.verbose = True
        self.editions = ["Épreuve du froid", "Au-delà des portes", "Au-delà des portes - Édition KS"]
        self.factions = ["ax", "br", "ly", "mu", "or", "yz"]
        self.df_my_cards = pd.DataFrame(columns=['Faction', 'Edition', 'Name', 'Number'])
        self.driver.get(self.url)

    def accept_cookies(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            accept_button = self.driver.find_element(By.ID, "didomi-notice-agree-button")
            accept_button.click()
            if self.verbose:
                print("Cookies accepted.")
        except Exception as e:
            print("No cookies popup found or an error occurred:", e)

    def login(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        # 1. Find the button using its class name and click it
        try:
            login_button = self.driver.find_element(By.CSS_SELECTOR, "button.hover\\:rounded-tl.hover\\:rounded-br.transition-all.duration-quicker.ease-out.clickable.flex.items-center.whitespace-nowrap.justify-center.font-sans.py-3.px-4.sm-bold-caps.gap-x-4.bg-primary-500.text-neutral-900.shadow-raised.hover\\:bg-primary-300.active\\:bg-primary-600.disabled\\:bg-system-disabled-300.disabled\\:text-system-disabled-400")
            login_button.click()
            if self.verbose:
                print("Login button clicked.")
        except Exception as e:
            print("Login button not found or an error occurred:", e)
        # 2. Find the username input field and type in the username
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            username_input = self.driver.find_element(By.ID, "username")
            username_input.send_keys(self._login)
            if self.verbose:
                print("Email input successful.")
        except Exception as e:
            print("Username input field not found or an error occurred:", e)
        # 3. Find the password input field and type in the password
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            password_input = self.driver.find_element(By.ID, "password")
            password_input.send_keys(self._mdp)
            if self.verbose:
                print("Password input successful.")
        except Exception as e:
            print("Password input field not found or an error occurred:", e)
        # 4. Find the login button and click it
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            continue_button = self.driver.find_element(By.CSS_SELECTOR, "button.hover\\:rounded-tl.hover\\:rounded-br.transition-all.duration-quicker.ease-out.clickable.flex.items-center.whitespace-nowrap.justify-center.font-sans.w-full.p-2.sm-bold-caps.gap-x-2.bg-primary-500.text-neutral-900.shadow-raised.hover\\:bg-primary-300.active\\:bg-primary-600.disabled\\:bg-system-disabled-300.disabled\\:text-system-disabled-400.h-14.font-sans")
            continue_button.click()
            if self.verbose:
                print("Continue button clicked.")
        except Exception as e:
            print("Continue button not found or an error occurred:", e)    

    def go_to_cards(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            cards_link = self.driver.find_element(By.CSS_SELECTOR, "a.hover\\:no-underline.px-3.py-2.rounded-lg.flex.flex-col.md\\:flex-row.md\\:gap-2.justify-center.items-center.relative.sm-regular.md\\:md-bold-caps.whitespace-nowrap.text-neutral-800.hover\\:bg-primary-400.hover\\:text-neutral-800.drop-shadow-none")
            cards_link.click()
            if self.verbose:
                print("Cards link clicked.")
        except Exception as e:
            print("Cards link not found or an error occurred:", e)
            
    def open_filters(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            filters_button = self.driver.find_element(By.CSS_SELECTOR, "button.hover\\:rounded-tl.hover\\:rounded-br.transition-all.duration-quicker.ease-out.clickable.flex.items-center.whitespace-nowrap.justify-center.font-sans.py-3.px-4.sm-bold-caps.gap-x-4.bg-primary-500.text-neutral-900.shadow-raised.hover\\:bg-primary-300.active\\:bg-primary-600.disabled\\:bg-system-disabled-300.disabled\\:text-system-disabled-400.h-fit")
            filters_button.click()
            if self.verbose:
                print("Filters button clicked.")
        except Exception as e:
            print("Filters button not found or an error occurred:", e)

    def filter_click_faction(self, faction):
        # faction = "ax" or "br" or "ly" or "mu" or "or" or "yz"
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            fa_ax_element = self.driver.find_element(By.CSS_SELECTOR, f"i.fa-kit.fa-{faction}.text-2xl")
            fa_ax_element.click()
            if self.verbose:
                print(f"\t{faction} element clicked.")
        except Exception as e:
            print(f"\t{faction} element not found or an error occurred:", e)
    
    def filter_click_rare(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            rare_label = self.driver.find_element(By.CSS_SELECTOR, "label.transition.rounded-full.py-1.px-3.cursor-pointer.xs-bold-caps.whitespace-nowrap.flex.items-center.border-\\[2px\\].border-transparent.peer-disabled\\:text-system-disabled-400.peer-disabled\\:bg-system-disabled-300.peer-focus\\:shadow-focus.text-neutral-700.bg-mercury-100.hover\\:bg-neutral-200.peer-checked\\:text-neutral-50.peer-checked\\:bg-neutral-500.peer-checked\\:border-primary-500[for='rarity-Rare-RARE']")
            rare_label.click()
            if self.verbose:
                print("\tRare label clicked.")
        except Exception as e:
            print("\tRare label not found or an error occurred:", e)

    def filter_click_edition_accordion(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            edition_div = self.driver.find_element(By.XPATH, "//div[contains(@class, 'flex justify-between items-center gap-3') and text()='Édition']")
            edition_div.click()
            if self.verbose:
                print("\tÉdition accordion clicked.")
        except Exception as e:
            print("\tÉdition accordion not found or an error occurred:", e)

    def filter_click_specific_edition(self, edition):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            epreuve_du_froid_element = self.driver.find_element(By.XPATH, f"//p[@class='sm-bold-caps' and text()='{edition}']")
            epreuve_du_froid_element.click()
            if self.verbose:
                print(f"\t{edition} element clicked (to remove it)")
        except Exception as e:
            print(f"\t{edition} element not found or an error occurred:", e)

    def filter_click_continue(self):
        time.sleep(random.uniform(self.QUICK_WAIT[0], self.QUICK_WAIT[1]))
        try:
            filter_button = self.driver.find_element(By.CSS_SELECTOR, "button.hover\\:rounded-tl.hover\\:rounded-br.transition-all.duration-quicker.ease-out.clickable.flex.items-center.whitespace-nowrap.justify-center.font-sans.p-2.sm-bold-caps.gap-x-2.bg-primary-500.text-neutral-900.shadow-raised.hover\\:bg-primary-300.active\\:bg-primary-600.disabled\\:bg-system-disabled-300.disabled\\:text-system-disabled-400.grow")
            filter_button.click()
            if self.verbose:
                print("\tContinue button clicked.")
        except Exception as e:
            print("\tContinue button not found or an error occurred:", e)

    def get_page_names_list(self):
        elements = self.driver.find_elements(By.CSS_SELECTOR, "p.lg-bold.text-center")
        return [element.text for element in elements]
    
    def get_page_numbers_list(self):
        divs = self.driver.find_elements(By.CSS_SELECTOR, "div.p-1.justify-center.items-center.gap-1.flex.font-bold.min-w-.text-neutral-50.min-w-6.min-h-6")
        number_of_each_cards = []
        for div in divs:
            number = div.text
            if not number.isdigit():
                if number == '':
                    number = '0'
                else:
                    number = number[0]
            number_of_each_cards.append(int(number))
        return number_of_each_cards
    
    def click_next_page_if_present(self):
        # Scroll down to the bottom of the page
        self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(0.5)
        try:
            next_page_button = WebDriverWait(self.driver, 4).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "a.min-w-\\[40px\\].min-h-\\[40px\\].flex.items-center.justify-center.gap-2.px-4.py-2\\.5.text-neutral-900.md-medium.rounded-lg.hover\\:bg-mercury-200.hover\\:no-underline.transition-colors.duration-quicker.ease-out[data-testid='next-page']"))
            )
            next_page_button.click()
            new_page_available = True
            if self.verbose:
                print(f"Next page button clicked")
        except Exception as e:
            self.driver.execute_script("window.scrollTo(0, 0);")
            time.sleep(0.5)
            new_page_available = False
            if self.verbose:
                print("Next page button not found in 5 seconds")
        return new_page_available

    def main(self):
        self.accept_cookies()
        self.login()
        self.go_to_cards()
        for i, faction in enumerate(self.factions):                     # for every faction
            self.open_filters()
            self.filter_click_faction(faction)
            if i > 0:   # click on the previous faction to unselect it
                self.filter_click_faction(self.factions[i-1])
            if i == 0:  # only for the first faction (then it stays selected)
                self.filter_click_rare()
            if i > 0:   # reclick on last edition
                self.filter_click_edition_accordion()
                self.filter_click_specific_edition(self.editions[-1])
            for i_edi, edition in enumerate(self.editions):             # for every edition
                if i_edi > 0:   # reclick on filters after first edition done
                    self.open_filters()
                    self.filter_click_edition_accordion()
                if (i == 0) and (i_edi == 0):   # only for the first faction and the first edition
                    self.filter_click_edition_accordion()
                self.filter_click_specific_edition(edition)
                if i_edi > 0:   # we also have to click on the previous edition to unselect it
                    self.filter_click_specific_edition(self.editions[i_edi-1])
                self.filter_click_continue()
                
                new_page_available = True   
                while new_page_available == True:
                    time.sleep(random.uniform(self.QUICK_WAIT[0]*2, self.QUICK_WAIT[1]*2))
                    # print(f'page {page}')
                    names_lst = self.get_page_names_list()
                    number_lst = self.get_page_numbers_list()
                    
                    # Add names and number into df_my_cards
                    new_rows = []
                    for card_name, card_number in zip(names_lst, number_lst):
                        new_rows.append({'Faction': faction, 'Edition': edition, 'Name': card_name, 'Number': card_number})
                    self.df_my_cards = pd.concat([self.df_my_cards, pd.DataFrame(new_rows)], ignore_index=True) 
                    
                    new_page_available = self.click_next_page_if_present()

        self.df_my_cards.to_csv("my_cards.csv", index=False)
        self.driver.quit()
```

```python
scrp = Scrapper('https://www.altered.gg/')
scrp.verbose = False
scrp.main()

df_my_cards_raw = scrp.df_my_cards.copy()

# save the dataframe to a csv file
current_date = datetime.now().strftime("%Y-%m-%d")
df_my_cards_raw.to_csv(f'my_cards_{current_date}.csv', index=False)
```

```python
df_my_cards = df_my_cards_raw.copy()
# populate set number
df_my_cards['set'] = df_my_cards['Edition'].apply(lambda x: 2 if x == "Épreuve du froid" else 1)
my_factions = ['ax', 'ly']

for faction in my_factions:
    print('\nJ\'AI BESOIN DE :')
    print(f'Faction: {faction.upper()} ---------------------------')
    for set in [2, 1]:
        print(f'\tSet: {set}')
        df_sub = df_my_cards[(df_my_cards['Faction'] == faction) & (df_my_cards['set'] == set)]
        
        # some card are alternatives arts, and so they appear twice in the list (I don't care so I gather the information)
        df_sub = df_sub.groupby(['Name'], as_index=False).agg({'Faction': 'first', 'Edition': 'first', 'Number': 'sum', 'set': 'first'})
        df_sub['Quantity desired'] = 3 - df_sub['Number'].astype(int)
        df_sub = df_sub.sort_values(by='Name')
        # print(df_sub[df_sub['Quantity desired'] > 0][['Name', 'Quantity desired']].to_string(index=False))
        for index, row in df_sub[df_sub['Quantity desired'] > 0].iterrows():
            print(f'{row["Quantity desired"]} - {row["Name"]}')
        print(f'total: {df_sub["Quantity desired"].sum()}')

other_factions = [faction for faction in scrp.factions if faction not in my_factions]
for faction in other_factions:
    print('\nJE ME SÉPARE DE :')
    print(f'Faction: {faction} ---------------------------')
    for set in [2, 1]:
        print(f'\tSet: {set}')
        df_sub = df_my_cards[(df_my_cards['Faction'] == faction) & (df_my_cards['set'] == set)]
        df_sub = df_sub.sort_values(by='Name')
        if set == 1:
            df_sub['Name'] = df_sub.apply(lambda x: x['Name'] + ' (KS)' if 'KS' in x['Edition'] and df_sub[df_sub['Name'] == x['Name']].shape[0] > 1 else x['Name'], axis=1)
        elif set == 2:
            df_sub['Name'] = df_sub.apply(lambda x: x['Name'] + '' if df_sub[df_sub['Name'] == x['Name']].shape[0] > 1 and x.name == df_sub[df_sub['Name'] == x['Name']].index[0] else x['Name'], axis=1)
        # print(df_sub[df_sub['Number'] > 0][['Name', 'Number']].to_string(index=False))
        for index, row in df_sub[df_sub['Number'] > 0].iterrows():
            print(f'{row["Number"]} - {row["Name"]}')
        print(f'total: {df_sub["Number"].sum()}')        
```
